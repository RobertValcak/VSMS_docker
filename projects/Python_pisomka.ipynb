4. Definujte funkciu, ktorá vráti menšie z 2 čísel ak su obe párne a väčšie ak čísla párne niesu.

```python
# Kod
a=2
b=9
def porovnaj(a,b):
    if a% 2==0 and b%2==0:
        return min(a,b)
    else:
        return max(a,b)
print(porovnaj(a,b))
```
